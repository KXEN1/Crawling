```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# -----------------------------
# 1. 브라우저 설정
# -----------------------------
options = Options()
# options.add_argument("--headless")  # 화면 안 띄우고 실행하려면 주석 해제
options.add_argument("--disable-gpu")
options.add_argument("start-maximized")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# -----------------------------
# 2. 사이트 메인 접속
# -----------------------------
main_url = "https://contents.premium.naver.com/willam/william/contents/251122004412980pn"
driver.get(main_url)
time.sleep(3)  # 페이지 렌더링 대기

# -----------------------------
# 3. 로그인 창 열기
# -----------------------------
driver.find_element(By.CSS_SELECTOR, ".user_sign_in").click()  # 로그인 버튼 클릭
time.sleep(2)  # 로그인 창 로딩 대기

# -----------------------------
# 4. 로그인 입력
# -----------------------------
driver.find_element(By.ID, "id").send_keys("아이디")        # 아이디
driver.find_element(By.ID, "pw").send_keys("비번")  # 비밀번호
driver.find_element(By.ID, "log.login").click()             # 로그인 버튼 클릭
time.sleep(5)  # 로그인 후 페이지 로딩 대기

# -----------------------------
# 5. 로그인 후 바로 크롤링
# -----------------------------
elements = driver.find_elements(By.XPATH, '//span[starts-with(@id, "SE-")]')
results = [el.text.strip() for el in elements if el.text.strip()]

print(f"총 {len(results)}개의 span 텍스트 추출")
for r in results[:5]:  # 샘플 출력
    print(r)

# -----------------------------
# 6. txt 파일로 저장
# -----------------------------
with open("extracted_spans.txt", "w", encoding="utf-8") as f:
    for line in results:
        f.write(line + "\n")

print("extracted_spans.txt 저장 완료")

driver.quit()
```
